```python
import math 
import numpy as np
import pandas as pd
def calcPower(power, frequency, amplitude):
    x = power[0] * (math.log(frequency)/math.log(10))
    y = power[1] * math.log(amplitude)
    z = power[2]
    ans = math.pow(10,(x+y+z))
    return ans

def calcTipMass(mass, frequency):
    x = 2 * math.pi * frequency
    y = math.pow(x,2)    
    ans = ((mass[0]/y)-(mass[1]/1000))*1000
    return ans

def calcDis(disp, frequency, amplitude):
    x = disp[0] * (math.log(frequency)/math.log(10))
    y = disp[1] * math.log(amplitude)
    z = disp[2]
    ans = math.pow(10,(x+y+z))
    return ans

def calcV(ocv, frequency, amplitude):
    x = ocv[0] * (math.log(frequency)/math.log(10))
    y = ocv[1] * math.log(amplitude)
    z = ocv[2]
    ans = math.pow(10,(x+y+z))
    return ans

def findPPA(ppa_table, Frequency, Amplitude):
    df = pd.DataFrame(columns=('device','power','mass','disp','voltage','cost'))
    for index, row in ppa_table.iterrows():
        ppa = index
        power = calcPower([row['p1'],row['p2'],row['p3']],Frequency,Amplitude)
        mass = calcTipMass([row['stiff'],row['eff']],Frequency)
        disp = calcDis([row['d1'],row['d2'],row['d3']],Frequency,Amplitude)
        if ((Frequency < row['maxF']) & (disp < row['maxD'])):
            voltage = calcV([row['ocv1'],row['ocv2'],row['ocv3']], Frequency, Amplitude)
            cost = row['cost']
            df.loc[len(df)]=[ppa,power,mass,disp,voltage,cost]
            #print(voltage)
    return df
            
       
PPA_1001_power=[ -1.6803, 0.7613, 3.3561]
PPA_1001_mass=[ 453.47, 0.97]
PPA_1001_voltage=[-1.1254, 0.3347, 3.1314]
PPA_1001_disp=[-1.2878, 0.2853, 2.9426]
PPA_1001_max = [109.05,24, 29.26]

PPA_1011_power=[-1.6297, 0.643, 3.3964]
PPA_1011_mass=[445.98, 0.52]
PPA_1011_voltage=[-1.1873, 0.28, 3.2183]
PPA_1011_disp=[-1.22, 0.299, 2.9999]
PPA_1011_max = [147.11, 20.5,13.55]

PPA_1012_power=[-1.1481, 0.6322, 2.2136]
PPA_1012_mass=[763.42, 0.74]
PPA_1012_voltage=[-0.9875, 0.2977, 2.6349]
PPA_1012_disp=[-0.9396, 0.312, 2.2875]
PPA_1012_max = [161.22,8,18.25]

PPA_1014_power=[-1.7049, 0.6504, 4.0159]
PPA_1014_mass=[2185.56, 0.35]
PPA_1014_voltage=[-1.1354, 0.247, 3.4845]
PPA_1014_disp=[-1.0047, 0.166, 2.385]
PPA_1014_max = [398.85,5,13.16]

PPA_1021_power=[-1.5603, 0.6245, 2.9523]
PPA_1021_mass=[260.91, 0.17]
PPA_1021_voltage=[-1.1393, 0.2912, 3.3163]
PPA_1021_disp=[-1.138, 0.3197, 2.7225]
PPA_1021_max = [197.17,11,10.86]

PPA_1022_power=[-1.4713, 0.6388, 3.0402]
PPA_1022_mass=[847.56, 0.51]
PPA_1022_voltage=[-0.9969, 0.2687, 3.3656]
PPA_1022_disp=[-1.2243, 0.3103, 2.7985]
PPA_1022_max = [204.37,7,10.16]

PPA_2011_power=[-1.7595, 0.64, 3.8367]
PPA_2011_mass=[574.12, 0.64]
PPA_2011_voltage=[-1.1631, 0.2492, 3.2667]
PPA_2011_disp=[-1.0911, 0.29, 2.5715]
PPA_2011_max = [150.62,15.5,20.03]

PPA_2014_power=[-1.7229, 0.6528, 4.2545]
PPA_2014_mass=[3864.59, 0.39]
PPA_2014_voltage=[-1.1308, 0.2105, 3.58]
PPA_2014_disp=[-0.9748, 0.2107, 1.9594]
PPA_2014_max = [499.72,5,19.36]

PPA_4011_power=[-1.646, 0.6961, 4.2065]
PPA_4011_mass=[4097.35, 1.23]
PPA_4011_voltage=[-1.1713, 0.3005, 3.389]
PPA_4011_disp=[-1.0887, 0.2531, 2.3701]
PPA_4011_max = [290.96,5,30.15]

PPA_1001 = PPA_1001_power + PPA_1001_mass + PPA_1001_voltage + PPA_1001_disp + PPA_1001_max
PPA_1011 = PPA_1011_power + PPA_1011_mass + PPA_1011_voltage + PPA_1011_disp + PPA_1011_max
PPA_1012 = PPA_1012_power + PPA_1012_mass + PPA_1012_voltage + PPA_1012_disp + PPA_1012_max
PPA_1014 = PPA_1014_power + PPA_1014_mass + PPA_1014_voltage + PPA_1014_disp + PPA_1014_max
PPA_1021 = PPA_1021_power + PPA_1021_mass + PPA_1021_voltage + PPA_1021_disp + PPA_1021_max
PPA_1022 = PPA_1022_power + PPA_1022_mass + PPA_1022_voltage + PPA_1022_disp + PPA_1022_max
PPA_2011 = PPA_2011_power + PPA_2011_mass + PPA_2011_voltage + PPA_2011_disp + PPA_2011_max
PPA_2014 = PPA_2014_power + PPA_2014_mass + PPA_2014_voltage + PPA_2014_disp + PPA_2014_max
PPA_4011 = PPA_4011_power + PPA_4011_mass + PPA_4011_voltage + PPA_4011_disp + PPA_4011_max
a = np.array([PPA_1001])
b = np.array([PPA_1011])
c = np.array([PPA_1012])
d = np.array([PPA_1014])
e = np.array([PPA_1021])
f = np.array([PPA_1022])
g = np.array([PPA_2011])
h = np.array([PPA_2014])
i = np.array([PPA_4011])

mat = np.concatenate((a,b,c,d,e,f,g,h,i),axis=0)
index = ['PPA_1001', 'PPA_1011', 'PPA_1012', 'PPA_1014', 'PPA_1021', 'PPA_1022', 'PPA_2011', 'PPA_2014', 'PPA_4011']
ppa_table = pd.DataFrame(mat, columns=['p1', 'p2', 'p3', 'stiff', 'eff', 'ocv1', 'ocv2', 'ocv3', 'd1', 'd2', 'd3','maxF','maxD','cost'], index=index)
df_application = pd.read_csv("applications.csv")
```
